```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.ml.classification import RandomForestClassificationModel, GBTClassificationModel
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Start Spark session
spark = SparkSession.builder.appName("FinalModelAnalysis").getOrCreate()

# Load predictions
predictions = spark.read.option("header", True).option("inferSchema", True).csv("final_predictions_csv")

# Cast to correct types
predictions = predictions.withColumn("traffic_label", col("traffic_label").cast("double")) \
                         .withColumn("prediction", col("prediction").cast("double"))

# Evaluation
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(labelCol="traffic_label", predictionCol="prediction")

accuracy = evaluator.setMetricName("accuracy").evaluate(predictions)
f1 = evaluator.setMetricName("f1").evaluate(predictions)
precision = evaluator.setMetricName("weightedPrecision").evaluate(predictions)
recall = evaluator.setMetricName("weightedRecall").evaluate(predictions)

print("âœ… Classification Report:")
print(f"Accuracy :  {accuracy:.4f}")
print(f"F1 Score :  {f1:.4f}")
print(f"Precision:  {precision:.4f}")
print(f"Recall   :  {recall:.4f}")

# Confusion matrix
rdd = predictions.select("prediction", "traffic_label").rdd.map(lambda row: (row[0], row[1]))
metrics = MulticlassMetrics(rdd)
cm = metrics.confusionMatrix().toArray()

print("\nðŸ“Š Confusion Matrix:")
print(cm)

# Plot Confusion Matrix
plt.figure(figsize=(5, 4))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ["No Traffic", "Traffic"])
plt.yticks(tick_marks, ["No Traffic", "Traffic"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.grid(False)
plt.show()
```

```python
from pyspark.ml.tuning import CrossValidatorModel

# Load the full CrossValidator model (it wraps the RF/GBT inside)
model_path = "final_rf_model"
cv_model = CrossValidatorModel.load(model_path)

# Extract the best model (PipelineModel)
pipeline_model = cv_model.bestModel

# From that, extract the actual RandomForest model
rf_model = pipeline_model.stages[-1]  # last stage is usually the classifier

# Input features from VectorAssembler
input_features = ["Month", "Day", "day_of_week", "is_weekend", 
                  "Mean_Temp_C", "Total_Precip_mm", "temp_range", "day_of_year",
                  "season_vec_0", "season_vec_1", "season_vec_2", "season_vec_3"]

# Get feature importances from the RF model
importances = rf_model.featureImportances.toArray()

import pandas as pd
import matplotlib.pyplot as plt

# Combine and plot
importance_df = pd.DataFrame({"Feature": input_features, "Importance": importances})
importance_df = importance_df.sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df["Feature"], importance_df["Importance"])
plt.gca().invert_yaxis()
plt.xlabel("Importance Score")
plt.title("ðŸ“Š Feature Importance (Random Forest)")
plt.tight_layout()
plt.show()
```
